```python
import os
import re
import shutil
from collections import defaultdict

# Folder paths
source_folder = '../../images/upload/'  # Replace with your source folder path
destination_folder = '../../images/upload_new_1/'  # Replace with your destination folder path

# Ensure the destination folder exists
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Dictionary to track counts of files for each code_color combination
file_count = defaultdict(int)

# Function to standardize and clean up filenames
def clean_filename(filename):
    # Replace spaces and underscores with a single delimiter and remove extra characters
    cleaned = re.sub(r'[ _]+', '_', filename).replace('(', '').replace(')', '').lower()
    return cleaned

# Iterate over all files in the source folder
for filename in os.listdir(source_folder):
    # Check if it's a file
    if os.path.isfile(os.path.join(source_folder, filename)):
        # Split the filename based on underscores and spaces (using both as delimiters)
        parts = re.split(r'[ _]+', filename)

        if len(parts) >= 2:
            code = parts[0]  # First part is the code
            color = '_'.join(parts[1:-1])  # Combine the rest (excluding the last part, which is likely the position or suffix)
            ext = os.path.splitext(filename)[1]  # File extension (e.g., .png, .jpg)

            # Increment the count for each code_color combination
            file_count[(code, color)] += 1

            # Generate the new filename in the format code_color_count.extension
            new_filename = f"{code}_{color}_{file_count[(code, color)]}{ext}"

            # Clean up the new filename (standardize spaces and underscores)
            new_filename = clean_filename(new_filename)

            # Full path for source and destination files
            source_file = os.path.join(source_folder, filename)
            destination_file = os.path.join(destination_folder, new_filename)

            # Copy the file to the destination folder with the new name
            shutil.copy(source_file, destination_file)
            print(f"Renamed and copied '{filename}' to '{new_filename}'")

print(f"All files have been renamed and saved to {destination_folder}")
```

```python
import os
import re
import shutil
from collections import defaultdict

# Folder paths
source_folder = '../../images/upload/'  # Replace with your source folder path
destination_folder = '../../images/upload_new_1/'  # Replace with your destination folder path

# Ensure the destination folder exists
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Dictionary to track counts of files for each code_color combination
file_count = defaultdict(int)

# Function to clean up filenames (keeping code and color in uppercase)
def clean_filename(code, color, count, ext):
    # Keep both code and color in uppercase, clean and standardize the count
    cleaned_color = re.sub(r'[ _]+', '_', color).replace('(', '').replace(')', '').upper()
    new_filename = f"{code}_{cleaned_color}_{count}{ext}"
    return new_filename

# Iterate over all files in the source folder
for filename in os.listdir(source_folder):
    # Check if it's a file
    if os.path.isfile(os.path.join(source_folder, filename)):
        # Split the filename based on underscores and spaces (using both as delimiters)
        parts = re.split(r'[ _]+', filename)

        if len(parts) >= 2:
            code = parts[0].upper()  # First part is the code (force uppercase)
            color = '_'.join(parts[1:-1])  # Combine the rest (excluding the last part, which is likely the position or suffix)
            ext = os.path.splitext(filename)[1]  # File extension (e.g., .png, .jpg)

            # Increment the count for each code_color combination
            file_count[(code, color)] += 1

            # Generate the new filename while keeping both code and color in uppercase
            new_filename = clean_filename(code, color, file_count[(code, color)], ext)

            # Full path for source and destination files
            source_file = os.path.join(source_folder, filename)
            destination_file = os.path.join(destination_folder, new_filename)

            # Copy the file to the destination folder with the new name
            shutil.copy(source_file, destination_file)
            print(f"Renamed and copied '{filename}' to '{new_filename}'")

print(f"All files have been renamed and saved to {destination_folder}")
```
