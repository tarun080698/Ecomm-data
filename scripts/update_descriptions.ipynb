```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('../../data/product_data/Holiday_collection.csv')
# Remove duplicate rows based on a specific column (key), keeping the first occurrence
df_cleaned = df.drop_duplicates(subset='Code', keep='first')
# Save the cleaned data to a new CSV file
df_cleaned.to_csv('../../data/product_data/Holiday_collection_new.csv', index=False)

print("Duplicates removed and cleaned CSV saved as 'cleaned_output.csv'")
```

```python
import pandas as pd

# Read both CSV files into DataFrames
df_A = pd.read_csv('../../data/shopify_sheets/product_export_nova_vogue_10-29.csv')
df_B = pd.read_csv('../../data/product_data/Holiday_collection.csv')
# Convert 'Handle' column in File B to lowercase to match File A
df_B['Handle'] = df_B['Handle'].str.lower()
# Merge File A with File B to bring in the 'description' and 'price'
merged_df = pd.merge(
    df_A,
    df_B[['Handle', 'description', 'price']],
    on='Handle',
    how='left'
)

# Replace 'Body (HTML)' with 'description' from File B where available
# Add 'description' only for the first occurrence of each 'Handle'
merged_df['Body (HTML)'] = (
    merged_df.groupby('Handle')['description']
    .transform(lambda x: x.ffill().where(x.index == x.index.min(), ''))
)


# Update 'Variant Price' for all matching rows
merged_df['Variant Price'] = merged_df['price'].combine_first(merged_df['Variant Price'])

# Drop unnecessary columns from the merge
final_df = merged_df.drop(columns=['description', 'price'])

# Save the updated data to a new CSV file
final_df.to_csv('updated_file_A.csv', index=False)

print("File A updated successfully with prices and descriptions from File B.")
```
